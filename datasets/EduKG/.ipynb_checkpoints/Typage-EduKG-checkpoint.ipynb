```python
with open("ent2id.txt") as fin:
     rows = ( line.split('\t') for line in fin )
     ent2id = { row[0]:int(row[1]) for row in rows }
id2ent = {v: k for k, v in ent2id.items()}

with open("rel2id.txt") as fin:
     rows = ( line.split('\t') for line in fin )
     rel2id = { row[0]:int(row[1]) for row in rows }
id2rel = {v: k for k, v in rel2id.items()}

with open("type2id.txt") as fin:
     rows = ( line.split('\t') for line in fin )
     type2id = { row[0]:int(row[1]) for row in rows }
id2type = {v: k for k, v in type2id.items()}
```

```python
typid2entid = {}
entid2typid = {}


with open("train.txt", "r") as f:
    lines = f.readlines()

entid2typid = {}
for i, line in enumerate(lines):
    h = line.strip().split('\t')[0]
    t = line.strip().split('\t')[2]
    entid2typid[int(h)] = int(line.strip().split('\t')[8])
    entid2typid[int(t)] = int(line.strip().split('\t')[9])
```

```python
typid2entid = {}
for k,v in entid2typid.items():
    typid2entid.setdefault(v,[]).append(k)
```

```python
# This is for getting observed triples later (filtering scores)
from collections import defaultdict

def get_observed_triples():
    all_possible_hs = defaultdict(dict)
    all_possible_ts = defaultdict(dict)
    # For better performance, run this function locally and store results in pickle format, then save it definitely and load it here
    # (instead of running it every time)
    train = torch.as_tensor(genfromtxt('train2id.txt', delimiter='\t'), dtype=torch.int32)
    valid = torch.as_tensor(genfromtxt('valid2id.txt', delimiter='\t'), dtype=torch.int32)
    test = torch.as_tensor(genfromtxt('test2id.txt', delimiter='\t'), dtype=torch.int32)
    all_triples = torch.cat((train,valid,test))
    X = all_triples.detach().clone()
    for triple in range(X.shape[0]):
        h,r,t = X[triple][0].item(), X[triple][1].item(), X[triple][2].item()
        try:
            all_possible_ts[h][r].append(t)
        except KeyError:
            all_possible_ts[h][r] = [t]

    for triple in range(X.shape[0]):
        h,r,t = X[triple][0].item(), X[triple][1].item(), X[triple][2].item()
        try:
            all_possible_hs[t][r].append(h)
        except KeyError:
            all_possible_hs[t][r] = [h]

    all_possible_ts = dict(all_possible_ts)
    all_possible_hs = dict(all_possible_hs)
    return all_possible_hs, all_possible_ts
```

```python
all_possible_hs, all_possible_ts = get_observed_triples()
```

```python
with open('observed_heads.pkl', 'wb') as f:
    pickle.dump(all_possible_hs, f)
with open('observed_tails.pkl', 'wb') as f:
    pickle.dump(all_possible_ts, f)
```

```python
# domain-range for each relation in terms of admissible entity type
# for r in X_train[:,1].unique():
#     X_tmp = X_train[X_train[:,1]==r]
all_possible_ts = {}
all_possible_hs = {}
train = torch.as_tensor(genfromtxt('train2id.txt', delimiter='\t'), dtype=torch.int32)
valid = torch.as_tensor(genfromtxt('valid2id.txt', delimiter='\t'), dtype=torch.int32)
test = torch.as_tensor(genfromtxt('test2id.txt', delimiter='\t'), dtype=torch.int32)
X = torch.cat((train,valid,test))
for triple in range(X.shape[0]):
    h,r,t = X[triple][0].item(), X[triple][1].item(), X[triple][2].item()
    try:
        all_possible_ts[r].append(t)
    except KeyError:
        all_possible_ts[r] = [t]
    try:
        all_possible_hs[r].append(h)
    except KeyError:
        all_possible_hs[r] = [h]
        
dom_rel = {}
range_rel = {}
i=0
for r in train[:,1].unique():
    range_ent = all_possible_ts[r.item()]
    try:
        tmp = [entid2typid[ent] for ent in range_ent]
    except:
        print(i)
    range_type = list(set(itertools.chain.from_iterable(tmp)))
    range_rel[r.item()] = range_type
    
    
    dom_ent = all_possible_hs[r.item()]
    try:
        tmp2 = [entid2typid[ent] for ent in dom_ent]
    except:
        print(i)
    dom_type = list(set(itertools.chain.from_iterable(tmp2)))
    dom_rel[r.item()] = dom_type
    i+=1
```
