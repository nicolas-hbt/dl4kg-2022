```python
import torch.nn as nn
import torch
import numpy as np
import time
from collections import defaultdict
import pandas as pd
from torch.utils import data as torch_data
#from ..models import TransE, TransH, DistMult
import copy
import pickle
from numpy import genfromtxt
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

```python
train = pd.read_csv('train_t.csv', header=None, sep=',', names=['sid', 'pid', 'oid', 's','p','o','st','ot','stid','otid'])[0:]
valid = pd.read_csv('valid_t.csv', header=None, sep=',', names=['sid', 'pid', 'oid', 's','p','o','st','ot','stid','otid'])[0:]
test = pd.read_csv('test_t.csv', header=None, sep=',', names=['sid', 'pid', 'oid', 's','p','o','st','ot','stid','otid'])[0:]
```

```python
train = train.astype({"sid":"int","pid":"int","oid":"int","stid":"int","otid":"int"})
train.index = train.index.astype(int)
valid = valid.astype({"sid":"int","pid":"int","oid":"int","stid":"int","otid":"int"})
valid.index = valid.index.astype(int)
test = test.astype({"sid":"int","pid":"int","oid":"int","stid":"int","otid":"int"})
test.index = test.index.astype(int)
```

```python
test
```

```python
train2id = train[['sid', 'pid', 'oid']]
valid2id = valid[['sid', 'pid', 'oid']]
test2id = test[['sid', 'pid', 'oid']]

np.savetxt('train2id.txt', train2id.values, fmt='%d', delimiter='\t')
np.savetxt('valid2id.txt', valid2id.values, fmt='%d', delimiter='\t')
np.savetxt('test2id.txt', test2id.values, fmt='%d', delimiter='\t')
```

```python
df = pd.concat([train,valid,test])
```

```python
df.pid.unique()
```

```python
rel2id = {}
rel2id['author_in_affiliation'] = 0
rel2id['author_write_paper'] = 1
rel2id['paper_cite_paper'] = 2
rel2id['paper_in_domain'] = 3
rel2id['paper_in_venue'] = 4
```

```python
df = df[["sid","oid","s","o","st","ot","stid","otid"]]
ent2id = {}
type2id = {}
entid2typid = {}
for i, row in df.iterrows():
    if not row["s"] in ent2id:
        ent2id[row["s"]] = row["sid"]
    if not row["o"] in ent2id:
        ent2id[row["o"]] = row["oid"]
    if not row["st"] in type2id:
        type2id[row["st"]] = row["stid"]
    if not row["ot"] in type2id:
        type2id[row["ot"]] = row["otid"]
    if not row["sid"] in entid2typid:
        entid2typid[row["sid"]] = row["stid"]
    if not row["oid"] in entid2typid:
        entid2typid[row["oid"]] = row["otid"]
    
```

```python
entid2typid
```

```python
typid2entid = {}
for k,v in entid2typid.items():
    typid2entid.setdefault(v,[]).append(k)
```

```python
id2ent = {v: k for k, v in ent2id.items()}
id2rel = {v: k for k, v in rel2id.items()}
id2type = {v: k for k, v in type2id.items()}

with open('ent2id.pkl', 'wb') as f:
    pickle.dump(ent2id, f)
with open('id2ent.pkl', 'wb') as f:
    pickle.dump(id2ent, f)
    
with open('rel2id.pkl', 'wb') as f:
    pickle.dump(rel2id, f)
with open('id2rel.pkl', 'wb') as f:
    pickle.dump(id2rel, f)
    
with open('type2id.pkl', 'wb') as f:
    pickle.dump(type2id, f)
with open('id2type.pkl', 'wb') as f:
    pickle.dump(id2type, f)
    
with open('typid2entid.pkl', 'wb') as f:
    pickle.dump(typid2entid, f)
with open('entid2typid.pkl', 'wb') as f:
    pickle.dump(entid2typid, f)
```

```python
def get_observed_triples():
    all_possible_hs = defaultdict(dict)
    all_possible_ts = defaultdict(dict)
    # For better performance, run this function locally and store results in pickle format, then save it definitely and load it here
    # (instead of running it every time)
    train = torch.as_tensor(genfromtxt('train2id.txt', delimiter='\t'), dtype=torch.int32)
    valid = torch.as_tensor(genfromtxt('valid2id.txt', delimiter='\t'), dtype=torch.int32)
    test = torch.as_tensor(genfromtxt('test2id.txt', delimiter='\t'), dtype=torch.int32)
    all_triples = torch.cat((train,valid,test))
    X = all_triples.detach().clone()
    for triple in range(X.shape[0]):
        h,r,t = X[triple][0].item(), X[triple][1].item(), X[triple][2].item()
        try:
            all_possible_ts[h][r].append(t)
        except KeyError:
            all_possible_ts[h][r] = [t]

    for triple in range(X.shape[0]):
        h,r,t = X[triple][0].item(), X[triple][1].item(), X[triple][2].item()
        try:
            all_possible_hs[t][r].append(h)
        except KeyError:
            all_possible_hs[t][r] = [h]

    all_possible_ts = dict(all_possible_ts)
    all_possible_hs = dict(all_possible_hs)
    return all_possible_hs, all_possible_ts
```

```python
all_possible_hs, all_possible_ts = get_observed_triples()
```

```python
with open('observed_heads.pkl', 'wb') as f:
    pickle.dump(all_possible_hs, f)
with open('observed_tails.pkl', 'wb') as f:
    pickle.dump(all_possible_ts, f)
```
